# [Project Euler 206 - Concealed Squares](https://projecteuler.net/problem=206)

```python
import re
```

```python
def p206():
    n = isqrt(10^19) // 100
    while True:
        if re.match(r"1.2.3.4.5.6.7.8.9.0", str((n * 100 + 30)^2)):
            return n
        n += 1
```

```python
p206()
```
