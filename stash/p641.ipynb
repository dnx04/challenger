```sage
def iroot(x, n): 
    return ZZ(x).nth_root(n, truncate_mode=1)[0]

P = Primes()
def count_pair(limit, diff):
    res = 0
    for p in P:
        if p < limit // p + 1:
            res += prime_pi(limit // p + 1) - prime_pi(p)
            if p < diff and diff <= limit // p + 1:
                res -= 1
        else:
            break
    return res
```

```sage
def p641(limit):
    cand = [6 * k + 1 for k in range(1, 19)]
    ans = 0
    for c in cand:
        for p in P:
            if p^(c - 1) <= limit:
                ans += 1
            else:
                break
        if c == 25:
            ans += count_pair(iroot(limit, 4) + 1, 0)
        if c == 49:
            ans += count_pair(iroot(limit, 6) + 1, 0)
        if c == 91:
            for p in P:
                if p^16 <= limit:
                    ans += count_pair(iroot(limit // p^16, 2), p)
                else:
                    break
        if c == 55:
            for p in P:
                if p^4 * 2^10 > limit:
                    break
                for q in P:
                    if p == q:
                        continue
                    if p^4 * q^10 > limit:
                        break
                    ans += 1
                        
        if c == 85:
            for p in P:
                if p^4 * 2^16 > limit:
                    break
                for q in P:
                    if p == q:
                        continue
                    if p^4 * q^16 > limit:
                        break
                    ans += 1
    return ans
```

```sage
p641(10^8)
```
