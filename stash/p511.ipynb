```python
from functools import cache
```

```python
def p511(n, k):
    div = divisors(n)
    MOD = 10^9
    @cache
    def rec(sz, rem):
        ans = 0
        if sz == 0:
            if rem == 0:
                return 1
            else:
                return 0
        if sz % 2 == 0:
            for i in range(k):
                ans += rec(sz / 2, i) * rec(sz / 2, (rem - i + k) % k)
                ans %= MOD
        else:
            for d in div:
                ans += rec(sz - 1, (rem - d + k) % k)
                ans %= MOD
        return ans
    return rec(n, ((-n + k) % k + k) % k)
```
