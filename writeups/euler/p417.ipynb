# [Project Euler 417 - Reciprocal cycles II](https://projecteuler.net/problem=417)

```python
def cal(i):
    return (1 / i).period()
```

```python
%%cython

def p417(n):
    cdef long long ans = 0
    cdef int i
    for i in range(3, n + 1):
        tmp = i
        while tmp % 2 == 0:
            tmp /= 2
        while tmp % 5 == 0:
            tmp /= 5
        if tmp != 1:
            ans += cal(n)
    return ans
```

```python
p417(10^6)
```
