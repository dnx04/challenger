# Project Euler - 795

Bruteforce and save result along the way maybe enough.

```python
def p795(n):
    ans = 0
    for i in range(1, n + 1):
        if i % 2 == 0:
            ans += (n - i + 1) * gcd(n, i * i)
        else:
            ans -= (n - i + 1) * gcd(n, i * i)
        print(i, ans) 
    return ans
```

Let's verify the answer.

```python
assert p795(1234) == 2194708
p795(12345678)
```
