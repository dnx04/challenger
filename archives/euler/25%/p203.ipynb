# Project Euler - 203

Bruteforcing through all combinations is enough.

```python
def p203(x):
    s = set()
    for n in range(x):
        for k in range(n // 2 + 1):
            nCk = binomial(n, k)
            if squarefree_part(nCk) == nCk:
                s.add(nCk)
    return sum(s)
```
