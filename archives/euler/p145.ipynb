# [Project Euler 145 - How many reversible numbers are there below one-billion?](https://projecteuler.net/problem=145)

```python
def check_odd(s):
    for i in range(0, len(s)):
        if int(s[i]) % 2 == 0:
            return False
    return True

def p145(n):
    ans = 0
    for i in range(1, 5):
        b = 10^i + 1
        for j in range(1, b - 1):
            if j * b > 2 * n:
                break
            if check_odd(str(j)): 
                print(j, b)
                ans += j - j // 10
            j += 1
    return ans
```
