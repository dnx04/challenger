# [Project Euler 152 - Writing 1/2 as a sum of inverse squares](https://projecteuler.net/problem=152)

```sage
from functools import cache
```

```sage
LIM = 45
MAX_DIFF = sum([QQ(1/i^2) for i in range(2, LIM + 1)]) - QQ(1 / 2)

@cache
def rec(n, diff):
    if n > LIM:
        return 1
    res = 0
    if diff + QQ(1/n^2) <= MAX_DIFF:
        res += rec(n + 1, diff + QQ(1/n^2))
    res += rec(n + 1, diff)
    return res
```

```sage
rec(2, 0)
```
