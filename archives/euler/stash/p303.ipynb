# Project Euler - 303



```python
def p303(n):
    LIM = 20
    ans = 0
    for i in range(3, n + 1):
        pick = +Infinity
        dp = [[+Infinity for c in range(i)] for r in range(LIM)]
        dp[0][0] = 0
        print(i)
        for r in range(LIM - 1):
            for c in range(i):
                if dp[r][c] != +Infinity:
                    for d in [1, 2, 0]:
                        nxt = (c * 10 + d) % i
                        dp[r + 1][nxt] = min(dp[r + 1][nxt], dp[r][c] * 10 + d)
                        print(r + 1, nxt, dp[r + 1][nxt], r, c, dp[r][c])
                        if dp[r + 1][nxt] > 0 and nxt == 0:
                            pick = min(pick, dp[r + 1][nxt])        
        ans += pick // i
    return ans
```
