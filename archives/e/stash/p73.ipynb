# Project Euler 73 - Counting Fractions In A Range

The example sequence in the problem statement is a Farey sequence $F_n$.

http://people.csail.mit.edu/mip/papers/farey/talk.pdf

```python
def p73(n):
    f = [QQ(1/3), QQ(1/2)]
    for step in range(4, n + 1):
        pending = []
        for i in range(len(f) - 1):
            numer = f[i].numer() + f[i + 1].numer()
            denom = f[i].denom() + f[i + 1].denom()
            g = gcd(numer, denom)
            if g != 1:
                continue
            if denom <= step:
                pending.append(QQ(numer / denom))
        f = merge_increasing_sets(f, pending)
        print(step, len(f))
    return len(f) - 2
```

```python
p73(8)
p73(12000)
```
